# **Feature Encoding:**

**Label Encoder with Sklearn**

```python
List =['a','b','k','m','l','o','d','x','s']

data = pd.DataFrame(List,columns= ['Letters'])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['Enc'] = le.fit_transform(data['Letters'])
data
```

**Label Encoding created func:**

```python
import numpy as np
List =['c','h','vc','vh']
NewList = np.random.choice(List , size=20)

data = pd.DataFrame(List , columns =['Climate'])
data
```

```python
def ClimateOrder(i):
  if i =='vh':
    return 0
  elif i =='h':
    return 1
  elif i =='c':
    return 2
  elif i =='vc':
    return 3

data['Enc']= data['Climate'].apply(ClimateOrder)
data
```

```python
data.to_csv('new.csv')
data.to_excel('b.xlsx')
```
